```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

 Load iris dataset

```python
iris=load_iris()
x=iris.data
y=iris.target
print('The features names is : ',iris.feature_names)
print('The Shape of the features is :' ,x.shape)
print('The Shape of the classes is :' ,y.shape)
```

Standerdize the data

```python
#StandardScaler to make the distribution's mean zero and variance equals to 1
X_std=StandardScaler().fit_transform(x)
print(X_std)
```

Apply LDA ( reduce the number of columns from 4 to 2 )

```python
LDA=LinearDiscriminantAnalysis(n_components=2)
X_lda=LDA.fit_transform(X_std,y)
print('the dataset shape after appling LDA is : ',X_lda.shape)
print('the dataset shape before appling LDA is : ',X_std.shape)
print(X_lda)
```

```python
plt.figure(figsize=(10,5));
plt.xlabel('LDA 1');
plt.ylabel('LDA 2');
plt.scatter(X_lda[:,0],X_lda[:,1],c=y);
```
