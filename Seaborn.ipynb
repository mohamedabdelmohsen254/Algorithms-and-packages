**Viloin Plot**

A violin plot is more informative than a plain box plot. While a box plot only shows summary statistics such as mean/median and interquartile ranges, the violin plot shows the full distribution of the data.

while a box plot is going to correspond to data points a violin plot uses the kde estimation of our data points to create these plots 

```python
sns.violinplot(x='day',y='total_bill',data=tips_df,hue='sex')
```

```python
#split's going to allow you to compare how the categories compare to each other 
sns.violinplot(x='day',y='total_bill',data=tips_df,hue='sex',split=True)
```

**Strip plots**

A strip plot is a graphical data anlysis technique for summarizing a univariate data set.

what a strip plot does is it's going to draw a scatter plot representing all the different data points where one variable is categorical

```python
plt.figure(figsize=(8,5))

sns.stripplot(x='day',y='total_bill',data=tips_df)
```

```python
plt.figure(figsize=(8,5))

sns.stripplot(x='day',y='total_bill',data=tips_df,hue='sex')
```

**swarm plot**

## Matrix Plots

**Heatmaps**

A heatmap contains values representing various shades of the same colour for each value to be plotted.

Usually the darker shades of the chart represent higher values than the lighter shade.

[Matplotlib Colormaps](https://matplotlib.org/stable/tutorials/colors/colormaps.html)


```python
plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)

crash_matrix=crash_df.corr()
#annot will show the numbers inside each colored box
sns.heatmap(crash_matrix,annot=True)
```

**Pair Grid**

```python
#we will map to plt.scatter where all plots will be a scatterplots
#sns.pairplot(iris)
iris=sns.load_dataset('iris')
iris_grid=sns.PairGrid(iris)
iris_grid.map(plt.scatter)
```

```python
#This will print the same result that we can get from a pairplot
#sns.pairplot(iris)
iris=sns.load_dataset('iris')
iris_grid=sns.PairGrid(iris,hue='species')
iris_grid.map(plt.scatter)
```

```python
#what if we want to plot a histogram in the diagonal?
iris=sns.load_dataset('iris')
iris_grid=sns.PairGrid(iris,hue='species')
iris_grid.map(plt.scatter)
iris_grid.map_diag(plt.hist)
```

```python
#what if we want to plot the scatter plots off the diagonal?
iris=sns.load_dataset('iris')
iris_grid=sns.PairGrid(iris,hue='species')
iris_grid.map_diag(plt.hist)
iris_grid.map_offdiag(plt.scatter)
```

```python
#what if we want to plot scatter in the upper and different plot in the lower?
iris=sns.load_dataset('iris')
iris_grid=sns.PairGrid(iris,hue='species')
iris_grid.map_diag(plt.hist)
iris_grid.map_upper(plt.scatter)
iris_grid.map_lower(sns.kdeplot)
```

```python
#we can define variables x and y for custom plot
iris_grid=sns.PairGrid(iris,hue='species',
                       x_vars=['sepal_length','sepal_width'],
                      y_vars=['petal_length','petal_length'])
iris_grid.map(plt.scatter)
iris_grid.add_legend()
```
