```python
#importing the libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

 Load iris dataset

```python
iris=load_iris()
x=iris.data
y=iris.target
print('The features names is : ',iris.feature_names)
print('The Shape of the features is :' ,x.shape)
print('The Shape of the classes is :' ,y.shape)
```

Standerdize the data

```python
#StandardScaler to make the distribution's mean zero and variance equals to 1
X_std=StandardScaler().fit_transform(x)
print(X_std)
```

Apply PCA ( reduce the number of columns from 4 to 2 )

```python
pca_model=PCA(n_components=2)
X_pca=pca_model.fit_transform(X_std)
print('the dataset shape after appling PCA is : ',X_pca.shape)
print('the dataset shape before appling PCA is : ',X_std.shape)
print(X_pca)
```

visualize the data

```python
plt.figure(figsize=(10,5))
plt.xlabel('prinicpal componant 1')
plt.ylabel('prinicpal componant 2')
plt.scatter(X_pca[:,0],X_pca[:,1],c=y)
```
